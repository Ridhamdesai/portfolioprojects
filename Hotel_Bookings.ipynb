### Data Analysis and Visualization

```python
cancelled_percentage = df['is_canceled'].value_counts(normalize = True)
print(cancelled_percentage)

plt.figure(figsize = (5,4))
plt.title('Reservation Status')
plt.bar(['Not Cancelled','Cancelled'],df['is_canceled'].value_counts(), width = 0.7)
plt.show()
```

```python
plt.figure(figsize = (8,4))
ax1 = sns.countplot(x = 'hotel',hue ='is_canceled',data = df,palette = 'Blues')
legend_labels,_ = ax1.get_legend_handles_labels()
labels = ['Not Canceled', 'Canceled']
ax1.legend(labels=labels, bbox_to_anchor=(1, 1))
plt.title('Reservation status in different hotels',size = 20)
plt.xlabel('Hotel')
plt.ylabel('Number of Reservations')
```

```python
resort_hotel=df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize = (20,8))
plt.title('Average Daily Rate in City and Resort Hotel',fontsize = 30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label = 'Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize = (15,8))
ax1 = sns.countplot(x = 'month',hue ='is_canceled',data = df)
legend_labels,_ = ax1.get_legend_handles_labels()
labels = ['Not Canceled', 'Canceled']
ax1.legend(labels=labels, bbox_to_anchor=(1, 1))
plt.title('Reservation status per month',size = 20)
plt.xlabel('Hotel')
plt.ylabel('Number of Reservations')
```

```python
plt.figure(figsize = (20,8))
plt.title('ADR per Month')
sns.barplot(x = 'month',y = 'adr',data = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
plt.legend()
plt.show()
```

```python
cancelled_hotel = df[df['is_canceled']== 1]
top_10_country = cancelled_hotel['country'].value_counts()[:10]
plt.figure()
plt.title('Top 10 countries with Highest Cancelation')
plt.pie(top_10_country,autopct = '%.2f',labels = top_10_country.index)
plt.show()


```

```python
cancelled_hotel_adr = cancelled_hotel.groupby('reservation_status_date')[['adr']].mean()
cancelled_hotel_adr.reset_index(inplace = True)
cancelled_hotel_adr.sort_values('reservation_status_date', inplace = True)

not_cancelled_hotel = df[df['is_canceled']== 0]
not_cancelled_hotel_adr = not_cancelled_hotel.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_hotel_adr.reset_index(inplace = True)
not_cancelled_hotel_adr.sort_values('reservation_status_date', inplace = True)

plt.figure(figsize = (20,6))
plt.title('Average Daily Rate',fontsize = 30)
plt.plot(cancelled_hotel_adr['reservation_status_date'],cancelled_hotel_adr['adr'],label = 'Cancelled')
plt.plot(not_cancelled_hotel_adr['reservation_status_date'],not_cancelled_hotel_adr['adr'],label = 'Not Cancelled')
plt.legend(fontsize = 20)
plt.show()
```
